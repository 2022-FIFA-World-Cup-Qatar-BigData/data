```python
dataset = df_nationality
dataset['total_points'] = np.nan
```

최종 dataset 합치는 작업

```python
# FIFA 15-23 player 최종본 
df_player22 = pd.read_excel("/content/drive/MyDrive/2022/BigData/FIFA 15-23.xlsx", sheet_name=8)
```

```python
df_player = df_player22
```

```python
# nationality_name, Model_Position로 group by 한 후, Position_Best_Rating 에 대해 rank 매기기 
df_player['rank_in_nation'] = df_player.groupby(['nationality_name','Model_Position'])['Position_Best_Rating'].rank(method='first',ascending=False)
```

```python
df_player = df_player.astype({'rank_in_nation':'int'})
df_player.head()
```

```python
df_player = df_player.replace({"Korea Republic":"South Korea"})
```

```python
df_player.loc[df_player['nationality_name'].str.contains('Korea')] # 값이 같은 경우 rank 
```

```python
df_player.loc[df_player['nationality_name'].str.contains('United')]
```

```python
dataset.loc[dataset['team'].str.contains('United States')]
```

```python
for index in dataset.index:
  team1 = dataset.loc[index, ['team']]

  team1 = team1[0]

  # position B (수비수) 4명 
  for i in range(1, 5):
    col1 = 'B'+str(i)+'_point'
    col2 = 'B'+str(i)+'_value'

    try:
      data = df_player['Position_Best_Rating'].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'B') & (df_player['nationality_name'] == team1 )]
      dataset.loc[index, [col1]] = float(data)
      data2 = df_player['value_eur'].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'B') & (df_player['nationality_name'] == team1)]
      dataset.loc[index, [col2]] = float(data2)
    except:
      continue

  # position M (미드필더) 4명 
  for i in range(1, 5):
    col1 = 'M'+str(i)+'_point'
    col2 = 'M'+str(i)+'_value'
    try:
      data = df_player['Position_Best_Rating'].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'M') & (df_player['nationality_name'] == team1 )]
      dataset.loc[index, [col1]] = float(data)

      data2 = df_player['value_eur'].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'M') & (df_player['nationality_name'] == team1)]
      dataset.loc[index, [col2]] = float(data2)

    except:
      continue
  # position (공격수) 2명 -> T
  for i in range(1, 3):
    col1 = 'T'+str(i)+'_point'
    col2 = 'T'+str(i)+'_value'
    try:
      data = df_player['Position_Best_Rating'].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'T') & (df_player['nationality_name'] == team1 )]
      dataset.loc[index, [col1]] = float(data)
      data2 = df_player['value_eur'].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'T') & (df_player['nationality_name'] == team1)]
      dataset.loc[index, [col2]] = float(data2)

    except:
      continue

  # # 골키퍼 
  col1 = 'K1_point'
  col2 = 'K1_value'
  try:
    data = df_player['Position_Best_Rating'].loc[(df_player['rank_in_nation'] == 1) & (df_player['Model_Position'] == 'K') & (df_player['nationality_name'] == team1 )]
    dataset.loc[index, [col1]] = float(data)
    data2 = df_player['value_eur'].loc[(df_player['rank_in_nation'] == 1) & (df_player['Model_Position'] == 'K') & (df_player['nationality_name'] == team1)]
    dataset.loc[index, [col2]] = float(data2)
  except:
    continue
```
